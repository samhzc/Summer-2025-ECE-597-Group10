```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
uvic_data=pd.read_csv('/content/drive/MyDrive/ECE 597 Capstone/uvic_data_cleaned.csv')
kaggle_data=pd.read_csv('/content/drive/MyDrive/ECE 597 Capstone/kaggle_data_cleaned.csv')
```

```python
kaggle_data.head(5)
```

```python
# Add a 'source' column to track the origin of each dataset
uvic_data['source'] = 'uvic'
kaggle_data['source'] = 'kaggle'

# Concatenate the dataframes
combined_data = pd.concat([uvic_data, kaggle_data], ignore_index=True)

# Shuffle the combined dataframe to uniformly mix the data
mixed_data = combined_data.sample(frac=1).reset_index(drop=True)

# Display the head of the mixed data to verify
print("Mixed Data Head:")
print(mixed_data.head())

# Display the distribution of sources to confirm uniform mixing
print("\nSource Distribution:")
print(mixed_data['source'].value_counts())
```

```python
#delete the source column
mixed_data = mixed_data.drop(columns=['source'])
```

```python
# Find the index of the first row where the 'label' column is 'spam'
try:
    first_spam_index = mixed_data[mixed_data['label'] == 'spam'].index[0]
    print(f"The index of the first spam label is: {first_spam_index}")
except IndexError:
    print("No spam labels found in the DataFrame.")
```
