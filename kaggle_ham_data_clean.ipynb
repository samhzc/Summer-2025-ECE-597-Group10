```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
```

```python
data_path = '/content/drive/MyDrive/ECE592B/emails.csv'
```

```python
# data_path = '/content/drive/MyDrive/ECE592B/emails.csv'
raw_data =  pd.read_csv(data_path)
```

```python
raw_data.head(5)
```

```python

# Now, 'message' column contains each email as a string
messages = raw_data['message'].dropna().tolist()

# data
def parse_email(raw_email):
    fields = {}
    patterns = {
        'Message-ID': r'Message-ID:\s*(<[^>]+>)',
        'Date': r'Date:\s*(.*)',
        'From': r'From:\s*([^\n]+)',
        'To': r'To:\s*([^\n]+)',
        'Subject': r'Subject:\s*([^\n]*)',
    }
    for key, pattern in patterns.items():
        match = re.search(pattern, raw_email)
        fields[key] = match.group(1).strip() if match else None
    # Try to extract the body (everything after the first empty line)
    body = re.split(r'\n\s*\n', raw_email, maxsplit=1)
    fields['Body'] = body[1].strip() if len(body) > 1 else ""
    return fields

parsed_emails = [parse_email(msg) for msg in messages]
df = pd.DataFrame(parsed_emails)
# print(df.head())
```

```python
# only need the body column, and store the result to data_clean
data_clean = df['Body']
```

```python
# add label
dataclean_labeled = pd.DataFrame({'text': data_clean, 'label': 'ham'})
dataclean_labeled.head(10)
```
